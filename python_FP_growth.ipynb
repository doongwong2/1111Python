```python
from collections import Counter

def createSet(l):
    ll=[]
    for i in l:
        ll.extend(i)
    yield list(ll)

lst=[line.split() for line in open("/mushroom.dat").readlines()]
list1=[]
for i in createSet(lst):
    list1.extend(i)
d=Counter()
d.update(list1)
d=dict(d)
print(d)
```

```python
import mlxtend
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

lst=[line.split() for line in open("/mushroom.dat").readlines()]
transEncoder = TransactionEncoder()
transEncoder_array = transEncoder.fit(lst).transform(lst)
transEncoder_array

#dataFrame = pd.DataFrame(transEncoder_array)
dataFrame = pd.DataFrame(transEncoder_array,columns = transEncoder.columns_)
dataFrame
```

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets = fpgrowth(dataFrame, min_support = 0.1, use_colnames = True, max_len = 1)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
assoRules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.8)
assoRules
```
